### Pivot allows you to transform and reshape data

```python
import pandas as pd
df = pd.read_csv('weather.csv')
df
```

```python
# if we want our index as date and columns to be city, then: 
df.pivot(index = 'date',columns='city')
```

```python
# for slicing purpose: [if we want only humidity]
df.pivot(index = 'date',columns='city', values='humidity')
```

```python
# We can also do:
df.pivot(index='humidity', columns='city')
```

## Pivot Table

### Pivot Table is used to summurize and aggregate data inside data frame

```python
df2 = pd.read_csv('weather_2.csv')
df2
# in this data frame i have data for same day for each city - one is for morning and one is for evening. 
```

```python
df2.pivot_table(index='city', columns=['date'])
# it's showing the average temperature and humidity per each date. 
# here we've aggregated our data to a meaningful output
```

### Defining aggregate functions 

```python
df2.pivot_table(index='city', columns=['date'], aggfunc='sum')
# it's using numpy functions here. 
```

```python
df2.pivot_table(index='city', columns=['date'], aggfunc='count')
# mean is default function here
```

### margins

```python
df2.pivot_table(index='city', columns=['date'], margins=True)
```

## Grouper in pivot table

```python
df= pd.read_csv('weather_3.csv')
df
```

```python
# if we want, average temperature for month of May and December:
```

```python
# first convert date column as date type
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index = pd.Grouper(freq='M',key='date'), columns='city')
```
