```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')  # This is a list of top rated movies available in internet
```

```python
movies.head()  # top 5 movies
```

```python
movies.shape  # to know how many rows and columns
```

```python
# We want movies whose duration is atleast 200
# here we'll create a list which contains True and False values. 
boolean_values = []
for length in movies.duration:
    if length >= 200:
        boolean_values.append('True')
    else:
        boolean_values.append('False')
```

```python
# Now convert the list into pandas Series
is_long = pd.Series(boolean_values)
is_long.head()  # if we don't make it series, then we can't pass it to our dataframe. 
```

### More simpler ideas to work with

```python
required_movies = movies.duration >= 200    # here result is stored in a series
# required_movies       # it contains series of True and False values. 
```

```python
movies[required_movies]       # it prints our required movies. 
```

### We can directly pass a relational operator to the column values

```python
movies[movies.duration >= 200]      # directly it prints our desired result. 
```

```python
# if we want only title of the movies which are longer duration than 200
movies[movies.duration >= 200].title        # here we are dealing with a dataframe. 
```

```python
# or we can do:
movies[movies.duration >= 200]['title']     # so we are selecting a column from a dataframe. 
```

### We can use loc() to select our required row and column

```python
movies.loc[movies.duration >= 200, 'title']    # loc function is to select data based on row and column
```
