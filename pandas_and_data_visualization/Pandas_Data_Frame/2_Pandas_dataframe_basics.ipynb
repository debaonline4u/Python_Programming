### Conditionally Select Data

```python
# give me all the rows in dataframe where my temperature >= 32
df[df.temperature >= 32]
```

## This is more powerful version like: [ Select windspeed, event from TABLE_NAME where temperature >= 32 ]

```python
# Give me windspeed and event in dataframe where my temperature >= 32
df[['windspeed','event']][df.temperature >= 32]
```

```python
# to find the row where my temperature was maximum
df[df.temperature == df.temperature.max()]
```

```python
# to find the same result, we can do the below syntax also
df[df.temperature == df['temperature'].max()]
```

```python
# to find the day and temperature when temperature was maximum
df[['day','temperature']][df.temperature == df['temperature'].max()]
```
