```python
import pandas as pd
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
g = df.groupby('city')
g
```

```python
# to Print g we have to do:
# city_df is the data frame corresponds to each city. 
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
# to get specific group data
g.get_group('mumbai')     # prints only mumbai data
```

#### this above work is same as SQL statement:
#### [SELECT * from city_data GROUP BY city HAVING city='mumbai']

```python
# Now we have to find maximum temperature in each city
g.max()   # this is so inconsistant  to use max()
```
