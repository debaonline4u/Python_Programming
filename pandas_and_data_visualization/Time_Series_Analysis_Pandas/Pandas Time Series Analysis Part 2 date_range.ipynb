### Here we're processing data which has no date column, we've to regenerate the date column for further procesing.

```python
import pandas as pd
df = pd.read_csv('aapl_no_dates.csv')
df.head()

# No date column here. It has int index. Now create a date column
```

```python
rng = pd.date_range(start='6/1/2017', end = '6/30/2017', freq='B')
rng

# Here we've created a date range for business date range, i.e. it excludes weekend holidays. 
```

```python
# to set this date to index. 

df.set_index(rng, inplace=True)
```

```python
df

# With date time index we can do a lot of things. 
# it's quite handy to do a lot of things. 
```

```python
## WE can do plotting according to Close volume

%matplotlib inline
df['Close'].plot()
```

```python
# we can do slicing
# if we want data for first 10 days

df['2017-06-01':'2017-06-10']
```

```python
df['2017-06-01':'2017-06-10'].Close.mean()

# first 10 days average. 
```

```python
# Suppose we want data on our weekend days also. 
# for that, let's pad the friday data to satureday and sunday. 

df.asfreq('D', method='pad')

# Now we've week end data too, and it's copied from Friday data. 
# study detail about asfreq() function 
```

```python
# we can show data for whole hour / each hour,
# though our dataset doesn't have hourly data, but wecan do it by asfreq() function. 

df.asfreq('H', method='pad')

# Now we've hourly data with data copied from last valid dataset. 
```

```python
## Suppose I have start date, and don't have end date, but I have period, let's say 72 days. 
# I want to generate new date range.

rng = pd.date_range(start='1/1/2017', periods=72, freq='B')
rng

# Here I got 72 business days. 

# this can be useful when we're writing some test data. 
```

```python
# Let's say we'll generate some random numbers 

import numpy as np
np.random.randint(1,10,len(rng))

# Here we got 72 random integers. 

# we can use it to generate a pandas Series. 
```

```python
ts = pd.Series(np.random.randint(1,10, len(rng)), index=rng)
ts.head(10)

# it's useful if we want some randome data or some sample of data. 
```
