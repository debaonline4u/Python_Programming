```python
# now our date column is a string. to make it as date format
df = pd.read_csv('aapl.csv', parse_dates=['Date'])
df.head()
# Now datatype of Date is changed to TimeStamp. 
```

```python
# to change the index to date column
df = pd.read_csv('aapl.csv', parse_dates=['Date'], index_col='Date')
df.head()
```

## Benefits of making Date as index and timeStamp format

### Retrieve Jan 2017 stock prices

```python
df['2017-01']
```

### Find Average stock of apple in month of January 2017

```python
df['2017-01'].Close.mean()
```

```python
# Price List of any given date
df['2017-01-03']
```

### WE can get the data for a given range too

```python
df['2017-01-07':'2017-01-01']
```

## 2) Resampling

### To retrieve monthly data on apple stock

```python
# select a column first and then resample to monthly frequency. 
df.Close.resample('M').mean()  # we can use count() and size() here
```

```python
%matplotlib inline
df.Close.resample('Q').mean().plot(kind='bar')  # quarterly bar chart
```
